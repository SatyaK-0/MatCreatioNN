```python
zip_file_path = "2019-MOFS.zip"
extract_folder = "2019-MOFS"
cif_folder = os.path.join(extract_folder, "structure_11660")  # Path to the CIF files
output_folder = os.path.join(extract_folder, "output")  # Path to the output folder
output_zip_path = "2019-MOFS-output.zip"  # Path to the output zip file

def zip_folder(folder_path, output_zip_path):
    with zipfile.ZipFile(output_zip_path, 'w', zipfile.ZIP_DEFLATED) as zipf:
        for root, dirs, files in os.walk(folder_path):
            for file in files:
                file_path = os.path.join(root, file)
                arcname = os.path.relpath(file_path, folder_path)
                zipf.write(file_path, arcname)

# Zip the output folder
zip_folder(output_folder, output_zip_path)
print(f"Zipped the output folder to: {output_zip_path}")

print("Processing complete!")
```

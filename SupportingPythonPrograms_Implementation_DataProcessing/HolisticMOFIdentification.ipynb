```python
import pandas as pd

# Load the CSV
df = pd.read_csv('MOF_Data Analysis.csv', header = None)  # Replace with your file name

# Get sets of unique values from each column
sets = [set(df[col].dropna().unique()) for col in df.columns]

# Find intersection (common values in all columns)
common_values = set.intersection(*sets)

print("Common values in all columns:", common_values)
```
